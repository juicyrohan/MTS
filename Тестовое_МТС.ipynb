 В МТС банке запустили аб-тест на странице, где можно оформить заявку на банковскую карту. В варианте А из 2574 клиентов, зашедших на страницу, на баннер нажали 350, в варианте Б из 2855 нажали 375. Какой вариант баннера предложить продакту и почему?


1. Находим величины конверсий CTR
 Вариант А
 conv_A = 350 / 2574 = 0.1360 = 13.6%
 Вариант B
 conv_B = 375 / 2855 =  0.1313 = 13.13%

2. Находим разницу между оценками конверсий. Она равна 13.6% – 13.13% = 0.47%.

3. Обозначим размеры выборок N_A = 2574, N_B = 2855
 Доверительный интервал (будем использовать 95% доверительный интервал) для разницы двух конверсий считается по следующей формуле:

 conv_A - conv_B ± 1.96 * sqrt(  (conv_A*(1 - conv_A)) / N_A  + (conv_B*(1 - conv_B)) / N_B  )
 0.136 - 0.1313 ± 1.96 * sqrt( 0.136 * (1 - 0.136) / 2574 + 0.1313 * (1 - 0.1313) / 2855 )
 0.0047 ± 1.96 * 0.0093

 В результате расчетов получаем интервал [-0.0134, 0.0228]

 Примечание: sqrt() - функция, извлекающая квадратный корень
4. Интервал содержит 0, поэтому мы вынуждены сказать, что статистически значимой разницы между результатами нет. Следовательно, на основании имеющихся данных конверсии в клик (CTR) в варианте А и варианте Б баннера неразличимы. То есть полученное изменение в конверсии вполне может быть случайным.
Поэтому мы можем предложить продакту любой из вариантов баннеров, объяснив, что с точки зрения статистики значимой разницы между ними нет, или предложить какой-либо другой фактор, по которому мы выберем баннер.

2. Для топ-100 наиболее платежеспособных клиентов за всю историю покупок посчитать траты помесячно


1.
SELECT id_client, SUM(sum_tran)
FROM transact
WHERE DATEDIFF(NOW(), tran_time) <= 30
GROUP BY id_client

2.
SELECT id_client, MONTH(tran_time) as month, YEAR(tran_time) as year,  SUM(sum_tran) as sum
FROM transact
GROUP BY id_client, YEAR(tran_time), MONTH(tran_time)
ORDER BY sum
LIMIT 100
